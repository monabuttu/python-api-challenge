```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint 

# Import API key
from config import api_key

# Output File (CSV)
output_data_file = "output_data/cities.csv"
```

## Generate Cities List

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
weatherdf = pd.DataFrame({"City":''},index=[np.arange(1,2000)])
weatherdf["Country Code"] = ""
weatherdf["Lat"] = ""
weatherdf["Lng"] = ""
weatherdf["Temperature (F)"] = ""
weatherdf["Humidity (%)"] = ""
weatherdf["Cloudiness (%)"] = ""
weatherdf["Wind Speed (mph)"] = ""
weatherdf.head()
```

```python
url = 'https://api.openweathermap.org/data/2.5/weather?lat='

Querycounter = 1

for index, row in weatherdf.iterrows(): 
    try:
        lat = np.random.uniform(low=-90.000, high=90.000)
        lng = np.random.uniform(low=-180.000, high=180.000)            
        query_url = url + str(lat) + '&lon='+ str(lng) +'&units=imperial' + '&appid=' + api_key
        response = requests.get(query_url).json()
        print(f'Query {Querycounter}')
        Querycounter+=1
        print(query_url)
        weatherdf.loc[index, "Lat"] = lat
        weatherdf.loc[index, "Lng"] = lng
        if response['name'] in weatherdf['City'].unique():
            print('City:' + response['name'] + ' Not Found')
            print('-----------------------------------------------------------------------------------------------------------')
            continue
        else:
            print('City:' + response['name'])
            weatherdf.loc[index, "City"] = response['name']          
        if len(weatherdf.loc[weatherdf['City']!="",:]) > 501:
            print('-----------------------------------------------------------------------------------------------------------')
            break
        weatherdf.loc[index, "Country Code"]= response['sys']['country']
        weatherdf.loc[index, "Temperature (F)"]=response['main']['temp']
        weatherdf.loc[index, "Humidity (%)"]=response['main']['humidity']
        weatherdf.loc[index, "Cloudiness (%)"]=response['clouds']['all']
        weatherdf.loc[index, "Wind Speed (mph)"]=response['wind']['speed']
        print('---------------------------------------------------------------------------------------------------------------')
    except:
        print(f'City for coordinates {lat,lng} not found. Searching for next City.')
```

```python
len(weatherdf.loc[weatherdf['City']!="",:])
```

```python
finalweatherdf = weatherdf.loc[weatherdf['City']!="",:].reset_index(drop=True)
finalweatherdf.drop(finalweatherdf.tail(1).index,inplace=True)
finalweatherdf.to_csv(output_data_file)
finalweatherdf.head()
```

#### Latitude vs. Temperature Plot

```python

x_limit = max(finalweatherdf['Lat']+5)


x_axis = finalweatherdf['Lat']


temp = finalweatherdf['Temperature (F)']


plt.scatter(x_axis, temp, marker="o", facecolors="pink", edgecolors="black",
            s=20, alpha=0.75)


plt.ylim(min(finalweatherdf['Temperature (F)'])-5, max(finalweatherdf['Temperature (F)'])+5)


plt.xlim(min(finalweatherdf['Lat'])-5, x_limit)


plt.title("City Latitude vs Temperature(F)")
plt.xlabel("City Latitude")
plt.ylabel("Temperature(F)")
plt.grid()

plt.savefig("../Images/Temperature.png")
plt.show()
```
